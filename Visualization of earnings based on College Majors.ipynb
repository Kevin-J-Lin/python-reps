```python
recent_grads.plot(x='Sample_size',y='Median',kind='scatter',title='Median vs Sample Size',figsize=(10,5))
```

Based on the following graphical analysis with scatter plots, students in more popular majors tend to make less money. The same observation can be made for students with majors that tend to have a female majority and majors with a large amount of full time positions available.

Based on the histogram analysis above, roughly 40% of majors are predominately male and 60% of majors are predominately female. The most common median salary falls between 36,000 and 38,000.

```python
from pandas.plotting import scatter_matrix
```

```python
sorted_grads_sharewomen = recent_grads.sort_values('ShareWomen')
```

```python
sorted_grads_sharewomen.head(10).plot.bar(x='Major', y='ShareWomen')
```

```python
sorted_grads_sharewomen.tail(10).plot.bar(x='Major', y='ShareWomen')
```

From a previous scatter plot, one could observe that there is a negative correlation between majors that are predominately female versus median salary. Observing the top ten majors that are predominately male, one can see that the majors are largely STEM majors. The difference between salaries of males and females might also be a factor or the gender gap in STEM fields.
